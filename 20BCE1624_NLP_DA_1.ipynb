1) Utilize Python NLTK (Natural Language Tool Kit) Platform and do the following. 
Install relevant Packages and Libraries 

• Explore Brown Corpus and find the size, tokens, categories,  
• Find the size of word tokens?  
• Find the size of word types?  
• Find the size of the category “government”  
• List the most frequent tokens  
• Count the number of sentences  

```python
import nltk

# Download the Brown Corpus
nltk.download('brown')

# Import the Brown Corpus
from nltk.corpus import brown

# Find the size of the Brown Corpus
size = len(brown.words())
print("Size of Brown Corpus: ", size)

# Find the size of word tokens
tokens = len(brown.words())
print("Size of word tokens: ", tokens)

# Find the size of word types
types = len(set(brown.words()))
print("Size of word types: ", types)

# Find the size of the category "government"
category_size = len(brown.words(categories=['government']))
print("Size of the category 'government': ", category_size)

# List the most frequent tokens
fd = nltk.FreqDist(brown.words())
print("Most frequent tokens: ", fd.most_common(10))

# Count the number of sentences
sentences = len(brown.sents())
print("Number of sentences: ", sentences)
```

b) POS Tagged corpus:  
->The NLTK library also includes a variety of POS-tagged text corpora, such as the Brown Corpus, which contains a collection of texts that have been manually POS-tagged. Here is an example of how to access and print the first 10 POS-tagged words from the "News" category of the Brown Corpus:

```python
import nltk
nltk.download('brown')
from nltk.corpus import brown

news = brown.tagged_words(categories=['news'])
print(news[:10])
```

```python
import nltk
nltk.download('treebank')
from nltk.corpus import treebank

tagged_words = treebank.tagged_words()
print(tagged_words[:10])
```

c) Parsed Corpus:  
The NLTK library also includes a variety of parsed text corpora, such as the dependency_treebank Corpus, which contains a collection of texts that have been manually parsed and annotated with semantic roles. Here is an example of how to access and print the first 10 parsed sentences from the "dependency_treebank" section of the dependency_treebank Corpus:

```python
import nltk
nltk.download('dependency_treebank')
from nltk.corpus import dependency_treebank

parsed_sents = dependency_treebank.parsed_sents()
for i, sent in enumerate(parsed_sents[:6]):
    print(f"sentence {i+1} : {sent}")
```

e) Spoken Language Corpus:  
The last of these corpora, udhr, contains the Universal Declaration of Human Rights in over 300 languages. The fileids for this corpus include information about the character encoding used in the file, such as UTF8 or Latin1. Let's use a conditional frequency distribution to examine the differences in word lengths for a selection of languages included in the udhr corpus. The output is shown in 1.2 (run the program yourself to see a color plot). Note that True and False are Python's built-in boolean values.

```python
from nltk.corpus import udhr
languages = ['Chickasaw', 'English', 'German_Deutsch','Greenlandic_Inuktikut', 'Hungarian_Magyar', 'Ibibio_Efik']
cfd = nltk.ConditionalFreqDist(
    (lang, len(word))
    for lang in languages
    for word in udhr.words(lang + '-Latin1'))
cfd.plot(cumulative=True)
```

f) Semantic Tagged Corpus:  
ne such corpus that is available in NLTK is the WordNet corpus. The WordNet corpus contains a collection of words and their senses, each sense is linked to a synset (synonym set) which is a group of words that have the same meaning. It is a lexical database for the English language, which is also used to tag words in other corpora such as SemCor.

```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet

words = list(wordnet.words())
for i, word in enumerate(words[:10]):
    for sense in wordnet.synsets(word):
        print(f"word {i+1} : {word} => sense: {sense}")
```

3) Create a text corpus with a minimum of 200 words (unique content). Implement the 
following text processing   
• Word segmentation  
• Sentence segmentation  
• Convert to Lowercase  
• Stop words removal  
• Stemming  
• Lemmatization  
• Part of speech tagger  

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

# Define the corpus
corpus = """The cat sat on the mat. The dog barked at the cat. The cat ran away.
The dog chased the cat but the cat was too fast. The cat hid under a bush. 
The dog searched for the cat but couldn't find it. Eventually, the cat came out 
and the dog went back to its owner. The end."""

# Word segmentation
words = nltk.word_tokenize(corpus)
print("Word segmentation: ", words)

# Sentence segmentation
sentences = nltk.sent_tokenize(corpus)
print("Sentence segmentation: ", sentences)

# Convert to lowercase
corpus = corpus.lower()
print("Lowercase: ", corpus)

# Stop words removal
stopwords = nltk.corpus.stopwords.words('english')
filtered_words = [word for word in words if word not in stopwords]
print("Stop words removal: ", filtered_words)

# Stemming
stemmer = nltk.stem.PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
print("Stemming: ", stemmed_words)

# Lemmatization
lemmatizer = nltk.stem.WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print("Lemmatization: ", lemmatized_words)

# Part of speech tagger
tagged_words = nltk.pos_tag(filtered_words)
print("Part of speech tagger: ", tagged_words)
```
